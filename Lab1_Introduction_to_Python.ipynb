```python
## 1. Confusion Matrix Metrics (Python as a Calculator)
TP = 50
FP = 8
TN = 32
FN = 10
# Total number of predictions
total_predictions = TP + FP + TN + FN
print("Total Predictions:", total_predictions)

# Accuracy calculation
accuracy = (TP + TN) / (TP + FP + TN + FN)
print("Accuracy:", accuracy)

# Precision calculation
precision = TP / (TP + FP)
print("Precision:", precision)

# Recall calculation
recall = TP / (TP + FN)
print("Recall:", recall)

# F1 Score calculation
f1_score = (2 * precision * recall) / (precision + recall)
print("F1 Score:", f1_score)
```

```python
## 2. Calculating R^2 (Coefficient of Determination)
# Given values
SSR = 420  # Sum of Squares due to Regression
SSE = 180  # Sum of Squares due to Error

# Total Sum of Squares
SST = SSR + SSE
print("Total Sum of Squares (SST):", SST)

# R-squared calculation
R_squared = SSR / SST
print("R-squared (R²):", R_squared)

# Interpretation of R-squared
if R_squared > 0.7:
    print("The model explains more than 70% of the variance.")
else:
    print("The model explains",R_squared*100,"% of the variance.")
```

```python
## 3. Mean Squared Error (MSE) Using Python as Calculator
# Squared errors for 5 observations
squared_errors = [2, 4, 1, 9, 4]
# Calculate MSE
MSE = sum(squared_errors) / len(squared_errors)
print("Mean Squared Error (MSE):", MSE)
```

```python
## 4: Cook’s Distance (Influential Observation)
# Given values
e_squared = 9
MSE = 3
p = 3
h_ii = 0.2
n = 25
# D_i = (e_i^2 / (p * MSE)) * (h_ii / (1 - h_ii)^2) # This line was causing the error
# First part of the formula
cooks_distance = (e_squared / (p * MSE))*(h_ii / (1 - h_ii) ** 2)
print("Cook's Distance:", cooks_distance)
# Threshold value
threshold = 4 / n
print("Threshold (4/n):", threshold)
# Influence decision
if cooks_distance > threshold:
    print("The observation is influential.")
else:
    print("The observation is not influential.")
```
